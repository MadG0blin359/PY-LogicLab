```python
age_q1 = emp_eda.loc['25%', 'Age']
age_q3 = emp_eda.loc['75%', 'Age']

sal_q1 = emp_eda.loc['25%', 'Salary']
sal_q3 = emp_eda.loc['75%', 'Salary']

age_iqr = age_q3 - age_q1
sal_iqr = sal_q3 - sal_q1

print("Age IQR: ", age_iqr)
print("Salary IQR: ", sal_iqr)
```

```python
age_upper_fence = age_q3 + (1.5 * age_iqr)
age_lower_fence = age_q1 - (1.5 * age_iqr)

sal_upper_fence = sal_q3 + (1.5 * sal_iqr)
sal_lower_fence = sal_q1 - (1.5 * sal_iqr)

print("Age Upper Fence: ", age_upper_fence)
print("Age Lower Fence: ", age_lower_fence)
print("Salary Upper Fence: ", sal_upper_fence)
print("Salary Lower Fence: ", sal_lower_fence)
```

```python
emp.loc[(emp['Salary'] > sal_upper_fence) | (emp['Salary'] < sal_lower_fence), 'Salary']
```

```python
emp.loc[(emp['Age'] > age_upper_fence) | (emp['Age'] < age_lower_fence), 'Age']
```

```python
emp['Age_ZScore'] = (emp['Age'] - emp_eda.loc['mean', 'Age']) / emp_eda.loc['std', 'Age']

display(emp[['Age', 'Age_ZScore']])
```

## Part 1

```python
cols = ['Age', 'Salary', 'Experience_Years', 'Performance_Score']
stats = emp[cols].agg([
    'mean',
    'median',
    'var',
    'std',
    'min',
    'max'
])

Q1 = emp[cols].quantile(0.25)
Q3 = emp[cols].quantile(0.75)
IQR = Q3 - Q1
Range = stats.loc['max'] - stats.loc['min']
Mode = emp[cols].mode().iloc[0]

iqr_series = pd.Series(IQR, name='IQR')
range_series = pd.Series(Range, name='Range')
mode_series = pd.Series(Mode, name='mode')

stats = pd.concat([stats, iqr_series.to_frame().T, range_series.to_frame().T, mode_series.to_frame().T])
stats.round(1)
```

## Part 2

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the numerical columns
cols = ['Age', 'Salary', 'Experience_Years', 'Performance_Score']

# Create a 2x2 grid for the four plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Histograms of Numerical Employee Features', fontsize=16)

# Flatten the axes array to easily iterate over it
axes = axes.flatten()

# --- Generate Histograms ---
for i, col in enumerate(cols):
    # We use .dropna() to ignore NaN values during plotting
    emp[col].dropna().hist(bins=20, ax=axes[i], color='teal', edgecolor='black')
    axes[i].set_title(f'{col} Distribution', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)

plt.show()
```

```python
# --- Set up the plotting area ---
sns.set_style("whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Boxplots for Outlier Visualization', fontsize=16)

# Flatten the axes array to easily iterate over it
axes = axes.flatten()

# --- Generate Boxplots ---
for i, col in enumerate(cols):
    # Use sns.boxplot, plotting against the y-axis (vertical orientation)
    # We use .dropna() to ignore NaN values during plotting
    sns.boxplot(y=emp[col].dropna(),
                ax=axes[i],
                color='lightcoral',
                width=0.4)
    axes[i].set_title(f'{col} Boxplot', fontsize=14)
    axes[i].set_ylabel(col, fontsize=12)
    axes[i].set_xticks([]) # Remove the x-ticks for cleaner vertical plot

plt.show()
```

```python
# --- Set up the plotting area ---
sns.set_style("whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Kernel Density Estimation (KDE) Plots', fontsize=16)

# Flatten the axes array to easily iterate over it
axes = axes.flatten()

# --- Generate KDE Plots ---
for i, col in enumerate(cols):
    # Use sns.kdeplot, which estimates the probability density function (PDF)
    # We use .dropna() to ignore NaN values
    sns.kdeplot(emp[col].dropna(),
                ax=axes[i],
                fill=True,
                color='darkblue',
                alpha=0.6,
                linewidth=2)

    axes[i].set_title(f'KDE for {col}', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Density', fontsize=12)

plt.show()
```

## Part 3

```python
from scipy.stats import skew

# Define the column to transform
col = 'Salary'

# --- 1. Apply Natural Log Transformation ---
emp['Salary_Log'] = np.log(emp[col])

# --- 2. Calculate Skewness Before and After ---
original_skew = skew(emp[col])
transformed_skew = skew(emp['Salary_Log'])

# --- 3. Plot Distributions ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Salary Distribution: Before and After Log Transformation', fontsize=16)

# Plot 1: Original Distribution (Highly Skewed)
sns.histplot(emp[col], bins=30, kde=True, ax=axes[0], color='lightcoral', edgecolor='black')
axes[0].set_title(f'Original Salary (Skew: {original_skew:.2f})', fontsize=12)
axes[0].set_xlabel('Salary (Original Scale)', fontsize=10)
axes[0].set_ylabel('Frequency', fontsize=10)

# Plot 2: Log Transformed Distribution (Normalized)
sns.histplot(emp['Salary_Log'], bins=30, kde=True, ax=axes[1], color='forestgreen', edgecolor='black')
axes[1].set_title(f'Log Transformed Salary (Skew: {transformed_skew:.2f})', fontsize=12)
axes[1].set_xlabel('Salary (Log Scale)', fontsize=10)
axes[1].set_ylabel('Frequency', fontsize=10)

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()

print(f"Original Salary Skewness: {original_skew:.4f}")
print(f"Log Transformed Salary Skewness: {transformed_skew:.4f}")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() # Initializing encoder

# Fit the encoder to the cleaned data and transform it
# The output is a new column with integer labels
emp['Gender_Encoded'] = le.fit_transform(emp['Gender'])

# Display
# .classes_ shows the original string values, and the index is the new integer
mapping = pd.DataFrame({
    'Original Label': le.classes_,
    'Encoded Value': range(len(le.classes_))
})

display(mapping)
print() # Newline
display(emp[['Gender', 'Gender_Encoded']])
```

```python
# pd.get_dummies() creates new columns (dummies) for each unique value in 'Department'.
# drop_first=True is highly recommended to avoid multicollinearity
# (the 'n-1' encoding trick).
emp_dummies = pd.get_dummies(
    emp['Department'],
    prefix='Department',
    drop_first=True,
    dummy_na=False # Assuming no missing values, but if there were, setting this to True
    # would create a 'Department_nan' column.
)

one_hot = pd.concat([emp, emp_dummies], axis=1)
display(one_hot['Department'].unique())
display(emp_dummies)
```
