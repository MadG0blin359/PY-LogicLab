# Part b

the crucial difference between correlation and causation. Just because two things move together doesn't mean one causes the other.

You can expect to see such strong, but ultimately meaningless, correlations in real datasets, especially when dealing with a large number of variables and looking for any kind of relationship. The more variables you have, the higher the chance of finding random correlations that appear significant.

To identify these "wrong" correlations, you need to:

- Apply domain knowledge: Does the correlation make sense in the real world? Is there a plausible reason for the two variables to be related causally?
- Consider confounding variables: Are there other factors that might be influencing both variables, creating the appearance of a direct relationship?
- Look for temporal precedence: Does the supposed cause happen before the supposed effect?
- Conduct controlled experiments: The best way to establish causation is through experiments where you manipulate one variable and observe the effect on another, while controlling for other factors.
- Be wary of data dredging (p-hacking): Avoid searching for correlations in a dataset without a pre-defined hypothesis. This increases the likelihood of finding statistically significant results purely by chance.

```python
import numpy as np
from scipy.stats import pearsonr

def calculate_correlation(arr1, arr2):
    # Ensure inputs are numpy arrays for consistent behavior
    arr1 = np.asarray(arr1)
    arr2 = np.asarray(arr2)

    # Check for constant arrays, which would result in NaN correlation
    if np.std(arr1) == 0 or np.std(arr2) == 0:
        return np.nan, np.nan, np.nan # Return NaN for correlation if one array is constant

    correlation, p_value = pearsonr(arr1, arr2)
    # Assuming degrees of freedom for two variables, n-2
    degrees_of_freedom = len(arr1) - 2
    return correlation, p_value, degrees_of_freedom

# Perform bootstrapping
n_iterations = 1000
correlations = []
n_samples = len(array_1)

for i in range(n_iterations):
    # Resample with replacement
    indices = np.random.choice(n_samples, size=n_samples, replace=True)
    array_1_sample = array_1.iloc[indices]
    array_2_sample = array_2.iloc[indices]

    # Calculate correlation for the resampled data
    correlation, _, _ = calculate_correlation(array_1_sample, array_2_sample)
    correlations.append(correlation)

# Remove any NaN values from correlations list that might arise from resampling constant arrays
correlations = [c for c in correlations if not np.isnan(c)]

# Calculate the 95% confidence interval using percentiles
confidence_interval = np.percentile(correlations, [2.5, 97.5])

# Print the results
print(f"Observed correlation coefficient (r): {correlation_spurious:.3f}")
print(f"95% Confidence Interval (bootstrapped): [{confidence_interval[0]:.3f}, {confidence_interval[1]:.3f}]")
```
