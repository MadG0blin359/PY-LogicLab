## Data Cleaning

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
df
```

```python
df.loc[0,"name"]
# df["name", 0].str.lower().str.title()
# df["name"] = df["name"].str.lower().str.title()
# df.loc[:, "name"] = df["name"].str.lower().str.title()
# df
```

```python
df["age"].unique()
```

```python
age_mean = df["age"].mean().round(0)
```

```python
df.loc[:, "age"] = df["age"].fillna(age_mean)
```

```python
df
```

```python
df["gender"].unique()
```

```python
df.loc[:, "gender"] = df["gender"].str.lower().str.title()
```

```python
df
```

```python
df["grade"]
```

```python
df["grade"].unique()
```

```python
df.loc[:, "grade"] = df["grade"].replace({'10th':'10', '11th':'11'})
```

```python
df
```

## Handle Missing Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# Select the rows that have at least one missing value
df[df.isnull().any(axis=1)].head()
```

```python
math_mean = df["math_score"].mean()
```

```python
df.loc[:, "math_score"] = df["math_score"].fillna(math_mean)
```

```python
df
```

```python
df["english_score"]
```

```python
df.loc[:, "english_score"] = pd.to_numeric(df["english_score"], errors='coerce')
```
